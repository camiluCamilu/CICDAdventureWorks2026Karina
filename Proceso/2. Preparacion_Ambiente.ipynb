#### 📁 1. External Locations - Estructura en ADLS

**¿Qué es un External Location?**

Un External Location en Databricks es un "puente" que conecta el catalog con carpetas en Azure (ADLS).

Básicamente: le dices a Databricks dónde están tus datos en Azure.

**¿Por qué cada carpeta separada?**

- ✅ **Orden**: Cada capa en su propia carpeta
- ✅ **Control**: Fácil aplicar permisos por carpeta
- ✅ **Escalabilidad**: Si crece, ya está organizado
- ✅ **Auditoría**: Puedes rastrear dónde vive cada dato

```sql
%python
dbutils.widgets.removeAll()
```

```sql
create widget text storageName default "adventureworksltadlsd";
```

```sql

DROP EXTERNAL LOCATION IF EXISTS`exlt-bronze`;
DROP EXTERNAL LOCATION IF EXISTS`exlt-silver`;
DROP EXTERNAL LOCATION IF EXISTS`exlt-golden`;
```

```sql
--Crear container delta-lake previamente en el datalake de la celda 2
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-mymetastore`
URL 'abfss://mymetastore@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential03`)
COMMENT 'Ubicación externa para las tablas delta del Data delta Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-bronze`
URL 'abfss://bronze@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential03`)
COMMENT 'Ubicación externa para las tablas delta del Data delta Lake en bronze';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-silver`
URL 'abfss://silver@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential03`)
COMMENT 'Ubicación externa para las tablas delta del Data delta Lake en silver';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-golden`
URL 'abfss://golden@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `credential03`)
COMMENT 'Ubicación externa para las tablas delta del Data delta Lake en golden';
```

```sql
%python

dbutils.fs.rm(f"abfss://bronze@adventureworksltadlsd.dfs.core.windows.net/",True);
dbutils.fs.rm(f"abfss://silver@adventureworksltadlsd.dfs.core.windows.net/",True);
dbutils.fs.rm(f"abfss://golden@adventureworksltadlsd.dfs.core.windows.net/",True);
```

```sql
DROP CATALOG IF EXISTS catadvwd CASCADE;
```

```sql
CREATE CATALOG IF NOT EXISTS catadvwd
MANAGED LOCATION 'abfss://mymetastore@${storageName}.dfs.core.windows.net/catadvwd/'
COMMENT 'Catalogo para la arquitectura medallion del ambiente de dev';
```

```sql
 

-- Schema Bronze (tablas delta sin procesar)
CREATE SCHEMA IF NOT EXISTS `catadvwd`.bronze
COMMENT 'Esquema para tablas bronze (sin procesar)'
;

-- Schema Silver (datos limpios y transformados)
CREATE SCHEMA IF NOT EXISTS `catadvwd`.silver
COMMENT 'Esquema para tablas silver (datos limpios)'
;

-- Schema Golden (datos para análisis)
CREATE SCHEMA IF NOT EXISTS `catadvwd`.golden
COMMENT 'Esquema para tablas golden (datos analíticos)'
;
```
