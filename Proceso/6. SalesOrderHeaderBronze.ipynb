```python
sales_order_header_schema = StructType(fields=[
    StructField('salesOrderId', IntegerType(), True),
    StructField('revisionNumber', IntegerType(), True),
    StructField('orderDate', TimestampType(), True),
    StructField('dueDate', TimestampType(), True),
    StructField('shipDate', TimestampType(), True),
    StructField('status', IntegerType(), True),
    StructField('onlineOrderFlag', IntegerType(), True),
    StructField('salesOrderNumber', StringType(), True),
    StructField('purchaseOrderNumber', StringType(), True),
    StructField('accountNumber', StringType(), True),
    StructField('customerId', IntegerType(), True),
    StructField('shipToAddressId', IntegerType(), True),
    StructField('billToAddressId', IntegerType(), True),
    StructField('shipMethod', StringType(), True),
    StructField('creditCardApprovalCode', StringType(), True),
    StructField('subTotal', DecimalType(10, 3), True),
    StructField('taxAmt', DecimalType(10, 3), True),
    StructField('freight', DecimalType(10, 3), True),
    StructField('totalDue', DecimalType(10, 3), True),
    StructField('comment', StringType(), True),
    StructField('rowguid', StringType(), True),
    StructField('modifiedDate', TimestampType(), True)
])
```

```python

catalogo = dbutils.widgets.get("catalogName")
esquema = dbutils.widgets.get("schemasName")
storageName = dbutils.widgets.get("storageName")

ruta = f"abfss://{esquema}@{storageName}.dfs.core.windows.net/SalesOrderHeader.csv"
```

```python
df_sales_order_header_final = spark.read \
    .option('header', True) \
    .schema(sales_order_header_schema) \
    .csv(ruta)
 
```

```python
display(df_sales_order_header_final)
```

```python
sales_order_header_selected_df = df_sales_order_header_final.select(
    col("salesOrderId"),
    col("revisionNumber"),
    col("orderDate"),
    col("dueDate"),
    col("shipDate"),
    col("status"),
    col("onlineOrderFlag"),
    col("salesOrderNumber"),
    col("purchaseOrderNumber"),
    col("accountNumber"),
    col("customerId"),
    col("shipToAddressId"),
    col("billToAddressId"),
    col("shipMethod"),
    col("creditCardApprovalCode"),
    col("subTotal"),
    col("taxAmt"),
    col("freight"),
    col("totalDue"),
    col("comment"),
    col("rowguid"),
    col("modifiedDate")
)
```

```python
sales_order_header_final_df = sales_order_header_selected_df.withColumn("createdDate", current_timestamp())
```

```python
display(sales_order_header_final_df)
```

```python
sales_order_header_final_df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("catadvwd.bronze.salesorderheader")
```
