```python
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("abfss://bronze@adventureworksltadlsd.dfs.core.windows.net/Customer.csv")

display(df)
```

```python
%sql 
--DROP TABLE IF EXISTS `${catalogName}`.`${schemasName}.customer
```

```python
customer_schema=StructType(fields=[
    StructField('customerId', IntegerType(), True),
    StructField('nameStyle', BooleanType(), True),
    StructField('title', StringType(), True),
    StructField('firstName', StringType(), True),
    StructField('middleName', StringType(), True),
    StructField('lastName', StringType(), True),
    StructField('suffix', StringType(), True),
    StructField('companyName', StringType(), True),
    StructField('salesPerson', StringType(), True),
    StructField('emailAddress', StringType(), True),
    StructField('phone', StringType(), True),
    StructField('createdDate', TimestampType(), True)
])

 
```

```python

catalogo = dbutils.widgets.get("catalogName")
esquema = dbutils.widgets.get("schemasName")
storageName = dbutils.widgets.get("storageName")

ruta = f"abfss://{esquema}@{storageName}.dfs.core.windows.net/Customer.csv"
```

```python
print (ruta)
```

```python
#ruta = "abfss://bronze@adventureworksltadlsd.dfs.core.windows.net/Customer.csv"
df_customer_final = spark.read\
.option('header', True)\
.schema(customer_schema)\
.csv(ruta)
```

```python
customer_selected_df= df_customer_final.select (col("customerId"), 
col("nameStyle"), 
col("title"), 
col("firstName"), 
col("middleName"), 
col("lastName"), 
col("suffix"), 
col("companyName"), 
col("salesPerson"), 
col("emailAddress"), 
col("phone"))
```

```python
display(customer_selected_df)
```

```python
customers_final_df = customer_selected_df.withColumn("createdDate", current_timestamp())
```

```python
display(customers_final_df)
```

```python
customers_final_df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("catadvwd.bronze.customers")
```
