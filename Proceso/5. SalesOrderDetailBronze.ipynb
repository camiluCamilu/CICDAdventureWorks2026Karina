```python
sales_order_detail_schema = StructType(fields=[
    StructField('salesOrderId', IntegerType(), True),
    StructField('salesOrderDetailId', IntegerType(), True),
    StructField('orderQty', IntegerType(), True),
    StructField('productId', IntegerType(), True),
    StructField('unitPrice', DecimalType(13, 2), True),
    StructField('unitPriceDiscount', DecimalType(13, 2), True),
    StructField('lineTotal', DecimalType(13, 2), True),
    StructField('rowguid', StringType(), True),
    StructField('modifiedDate', TimestampType(), True)
])
```

```python

catalogo = dbutils.widgets.get("catalogName")
esquema = dbutils.widgets.get("schemasName")
storageName = dbutils.widgets.get("storageName")

ruta = f"abfss://{esquema}@{storageName}.dfs.core.windows.net/SalesOrderDetail.csv"
```

```python
df_sales_order_detail_final = spark.read \
    .option('header', True) \
    .schema(sales_order_detail_schema) \
    .csv(ruta)
```

```python
#ruta = "abfss://bronze@adventureworksltadlsd.dfs.core.windows.net/SalesOrderDetail.csv"
df_sales_order_detail_final = spark.read \
    .option('header', True) \
    .schema(sales_order_detail_schema) \
    .csv(ruta)
```

```python

sales_order_detail_selected_df = df_sales_order_detail_final.select(
    col("salesOrderId"),
    col("salesOrderDetailId"),
    col("orderQty"),
    col("productId"),
    col("unitPrice"),
    col("unitPriceDiscount"),
    col("lineTotal"),
    col("rowguid"),
    col("modifiedDate")
)
```

```python
display(sales_order_detail_selected_df)
```

```python
sales_order_detail_final_df = sales_order_detail_selected_df.withColumn("createdDate", current_timestamp())
```

```python
display(sales_order_detail_final_df)
```

```python
sales_order_detail_final_df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("catadvwd.bronze.salesorderdetail")
```
